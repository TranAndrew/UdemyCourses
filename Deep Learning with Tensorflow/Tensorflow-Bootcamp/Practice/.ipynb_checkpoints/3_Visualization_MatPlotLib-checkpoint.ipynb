```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.arange(0,10)
x
```

```python
y = x**2
```

```python
plt.plot(x,y,'red')
plt.xlim(-1,10)
plt.ylim(-10,100)
plt.grid()
plt.title('Exponential Test Plot')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
mat = np.arange(0,100).reshape(10,10)
```

```python
## Add color gradient in matrix 
# check documentation 
plt.imshow(mat,cmap='cool')
```
